```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
num=np.random.randint(0, 5, 400)
num=pd.Series(num, name="risposte")
print(len(num))
```

```python
num2=np.random.randint(2, 7, 400)
num2=pd.Series(num2, name="risposte2")
print(len(num))
```

```python
plt.hist(num, 5)
```

```python
sns.set()

ax=sns.distplot(num, bins=5, kde=True, rug=True, hist=True, axlabel='pio', color='red', label='n.ro risposte', hist_kws={"color": "red", "edgecolor": "black"}, kde_kws={"shade": True})
bx=sns.distplot(num2, bins=5, kde=True, rug=True, hist=True, axlabel='pio2', color='blue', label='n.ro risposte2', hist_kws={"color": "blue", "edgecolor": "black"}, kde_kws={"shade": True})
# Plot formatting
plt.legend(prop={'size': 10})
plt.title('Life Expectancy of Two Continents')
plt.xlabel('Life Exp (years)')
plt.ylabel('Density')

```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="white", palette="muted", color_codes=True)
rs = np.random.RandomState(10)

# Set up the matplotlib figure
f, axes = plt.subplots(2, 2, figsize=(10, 10), sharex=True)
#sns.despine(left=True)

# Generate a random univariate dataset
d = rs.normal(size=100)

# Plot a simple histogram with binsize determined automatically
sns.distplot(d, kde=False, color="b", ax=axes[0, 0])

# Plot a kernel density estimate and rug plot
sns.distplot(d, hist=False, rug=True, color="r", ax=axes[0, 1])

# Plot a filled kernel density estimate
sns.distplot(d, hist=False, color="g", kde_kws={"shade": True}, ax=axes[1, 0])

# Plot a historgram and kernel density estimate
sns.distplot(d, color="m", ax=axes[1, 1])

plt.setp(axes, yticks=[])
plt.tight_layout()
```
